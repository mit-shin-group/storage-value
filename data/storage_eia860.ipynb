# Installed US Battery Storage

This notebook computes the numbers on installed US battery storage in Appendix A.1 of the manuscript.

```python
# import libraries
import pandas as pd
import numpy as np
```

```python
# Installed US electricity generation capacity
yr = 2025
nskip = 2
df_full = pd.read_excel("../data/eia-860/december_generator"+str(yr)+".xlsx", skiprows = nskip)
print("Total operating and standby capacity (GW): ", df_full[df_full["Status"].isin(["(OP) Operating", "'(SB) Standby/Backup: available for service but not normally used'"])]['Nameplate Capacity (MW)'].sum()/1000)
print("Total standby capacity (GW): ", df_full[df_full["Status"].isin(["'(SB) Standby/Backup: available for service but not normally used'"])]['Nameplate Capacity (MW)'].sum()/1000)
print("Total storage capacity (GW): ", df_full[(df_full["Status"].isin(["(OP) Operating"])) & (df_full["Technology"] == "Batteries")]['Nameplate Capacity (MW)'].sum()/1000)
print("Total utility owned storage capacity (GW): ", df_full[(df_full["Status"].isin(["(OP) Operating"])) & (df_full["Technology"] == "Batteries") & (df_full["Sector"] == "Electric Utility")]['Nameplate Capacity (MW)'].sum()/1000)
```
